```python
import numpy as np
import matplotlib.pyplot as plt
X=np.arange(start=-10,stop=10,step=0.5)
def f(x):
    return np.exp(x)/(1+np.exp(x))
```

```python
plt.plot(X,f(X))
```

```python
X=np.arange(start=-100,stop=100,step=0.5)
plt.plot(X,f(X))
```

```python
X=np.arange(start=-10,stop=10,step=0.5)
Y=f(X)
np.mean(Y)
```
